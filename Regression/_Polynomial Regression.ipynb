# #Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# #Importing the Dataset

```python
dataset=pd.read_csv("Position_Salaries.csv")
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
print(x)
print(y)
```

# #Training the Linear Regressor model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin1_reg=LinearRegression()
lin1_reg.fit(x,y)
```

# #Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
x_ploy = poly_reg.fit_transform(x)
lin2_reg = LinearRegression()
lin2_reg.fit(x_ploy, y)
```

# #Visualize the linear regression

```python
plt.scatter(x,y, color ="blue")
plt.plot(x, lin1_reg.predict(x), color = "green")
plt.title("Truth identification model using Linear regression")
plt.xlabel("Position")
plt.ylabel("Salaries")
plt.show()
```

# #Visualize the Polynomial regression

```python
plt.scatter(x,y, color="blue")
plt.plot(x, lin2_reg.predict(poly_reg.fit_transform(x)), color = "green")
plt.title("Truth identification model using Linear regression")
plt.xlabel("Position")
plt.ylabel("Salaries")
plt.show()
```

# #Visualization the polynomial regression result for higher dimension and smoother curve

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x,y, color = "blue")
plt.plot(x_grid, lin2_reg.predict(poly_reg.fit_transform(x_grid)), color = "green")
plt.title("Truth identification model using Linear regression")
plt.xlabel("Position")
plt.ylabel("Salaries")
plt.show()
```

```python
ploy_pred = lin2_reg.predict(poly_reg.fit_transform([[7.5]]))
print(ploy_pred)
```
